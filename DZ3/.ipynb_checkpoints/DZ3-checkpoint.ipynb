### **Задание 1** <a id='1'>
<a href="#up">наверх ↑</a>

*Импортируйте библиотеки pandas и numpy*

```python
import numpy as np
import pandas as pd
```

*Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и y из этих данных.*

```python
from sklearn.datasets import load_boston
boston = load_boston()

data = boston["data"]
feature_names = boston["feature_names"]
target = boston["target"]

X = pd.DataFrame(data, columns=feature_names)
y = pd.DataFrame(target, columns=["price"])
X
```

*Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.*

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

*Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.*

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

*Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.*

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred.shape
```

*Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.*

```python
from sklearn.metrics import r2_score
z1_result = r2_score(y_test, y_pred)
z1_result
```

### **Задание 2** <a id='2'>
<a href="#up">наверх ↑</a>

*Создайте модель под названием model с помощью RandomForestRegressor из модуля sklearn.ensemble.Сделайте агрумент n_estimators равным 1000,
max_depth должен быть равен 12 и random_state сделайте равным 42.*

```python

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

*Обучите модель на тренировочных данных аналогично тому, как вы обучали модель LinearRegression,
но при этом в метод fit вместо датафрейма y_train поставьте y_train.values[:, 0],
чтобы получить из датафрейма одномерный массив Numpymodel.fit(X_train, y_train.values[:, 0])*


```python
model.fit(X_train, y_train.values[:, 0])
```

*Сделайте предсказание на тестовых данных и посчитайте R2*

```python
y_pred = model.predict(X_test)
z2_result = r2_score(y_test, y_pred)
z2_result
```

*Сравните с результатом из предыдущего задания.Напишите в комментариях к коду, какая модель в данном случае работает лучше.*

```python
if z1_result > z2_result:
    print(f'Метод линейной регресии в данной задаче работает на {100*(z1_result-z2_result)/z2_result}% лучше')
elif z1_result < z2_result:
    print(f'Метод RandomForestRegressor в данной задаче работает на {round(100*(z2_result-z1_result)/z1_result,2)}% лучше')
else:
    print('Оба метода в данной задаче одинаково эффективны')
```

```python
df = pd.read_csv('creditcard.csv', delimiter=',')
```

```python
df['Class'].value_counts(normalize=True)
```

*Используя метод info, проверьте, все ли столбцы содержат числовые данные и нет ли в них пропусков.Примените следующую настройку, чтобы можно было просматривать все столбцы датафрейма:

```python
pd.options.display.max_columns = 100
df.info()
```

*Просмотрите первые 10 строк датафрейма df*

```python
df.head(10)
```

*Создайте датафрейм X из датафрейма df, исключив столбец Class.*

```python
X = df.drop('Class', axis=1)
```

*Создайте объект Series под названием y из столбца Class.*

```python
y = df['Class']
```

*Разбейте X и y на тренировочный и тестовый наборы данных при помощи функции train_test_split, используя аргументы: test_size=0.3, random_state=100, stratify=y.
У вас должны получиться объекты X_train, X_test, y_train и y_test.*


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.3, random_state=100, stratify=y)
```

*Просмотрите информацию о их форме.*

```python

print('X_train ', X_train.shape)
print('X_test ', X_test.shape)
print('y_train ', y_train.shape)
print('y_test ', y_test.shape)
```

*Для поиска по сетке параметров задайте такие параметры:
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]*


```python
parameters = [{
    'n_estimators': [10, 15], 
    'max_features': np.arange(3, 5), 
    'max_depth': np.arange(4, 7)
}]
```

*Создайте модель GridSearchCV со следующими аргументами:
estimator=RandomForestClassifier(random_state=100),
param_grid=parameters,
scoring='roc_auc',
cv=3.*

```python
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3
)
```

*Обучите модель на тренировочном наборе данных*

```python
clf.fit(X_train, y_train)
```

*Просмотрите параметры лучшей модели с помощью атрибута best_params_.*

```python
clf.best_params_
```

*Предскажите вероятности классов с помощью полученнной модели и метода predict_proba.*

```python
y_pred = clf.predict_proba(X_test)
```

*Из полученного результата (массив Numpy) выберите столбец с индексом 1 (вероятность класса 1) и запишите в массив y_pred_proba.*

```python
y_pred_proba = y_pred[:, 1]
y_pred_proba
```

*Из модуля sklearn.metrics импортируйте метрику roc_auc_score*

```python
from sklearn.metrics import roc_auc_score
```

*Вычислите AUC на тестовых данных и сравните с результатом,полученным на тренировочных данных, используя в качестве аргументов массивы y_test и y_pred_proba.*

```python
roc_auc_score(y_test, y_pred_proba)
```
